scraping data about hockey teams:
you have to scrap the first page table's data (Team name, year, wins, losses, win%, goals for(gf), golaagainst(ga))
from this website: https://www.scrapethissite.com/pages/forms/?page_num=1


```python
import requests
from bs4 import BeautifulSoup
url = "https://www.scrapethissite.com/pages/forms/?page_num=1"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('table', {'class': 'table'})
table
```

```python
table.text
```

```python
table.attrs
```

```python
headers = [header.text.strip() for header in table.find_all('th')]
headers
```

```python
with open("Nobel_laureates.csv", mode='w', encoding='utf-8') as fd:
    fd.write("Team_name, year, wins, losses, win%, goals_for_gf, golaagainst_ga\n")
    
    rows = table.find_all("tr")
    for row in rows[1:]:
        cells = row.find_all("td")
        if len(cells) == 0:
            continue
        Team_name = cells[0].text.strip()
        year = cells[1].text.strip()
        wins = cells[2].text.strip()
        losses = cells[3].text.strip()
        win = cells[4].text.strip()
        goals_for_gf = cells[5].text.strip()
        golaagainst_ga = cells[6].text.strip() if len(cells) > 6 else ""
    fd.write(f"{Team_name},{year},{wins},{losses},{win},{goals_for_gf},{golaagainst_ga}\n")
```
